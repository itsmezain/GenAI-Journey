## **Key Terms in Generative AI**  

### **1. Prompt Engineering (User Perspective)**  
ðŸ“Œ **Definition:**  
- The art of crafting inputs (prompts) to **get the best response** from an AI model.  
- Involves techniques like **few-shot learning, chain-of-thought prompting, and self-consistency**.  

ðŸ“Œ **Why It Matters?**  
- The **quality of the prompt** determines the **output quality**.  
- Used in **AI chatbots, content generation, creative writing, and automation**.  

ðŸ“Œ **Advanced Techniques:**  
- **Few-shot Prompting** â€“ Give examples to guide the model.  
- **Chain-of-Thought (CoT)** â€“ Make AI think step-by-step.  
- **Self-Consistency** â€“ Generate multiple responses and choose the best one.  

---

### **2. RLHF (Reinforcement Learning from Human Feedback) (Builder Perspective)**  
ðŸ“Œ **Definition:**  
- A method where a model is **fine-tuned using human feedback**.  
- AI generates responses â†’ Humans rank them â†’ AI learns **which responses are best**.  

ðŸ“Œ **Why It Matters?**  
- Used in **GPT-4, Claude, and Bard** to make them safer and more aligned.  
- Helps **reduce biases, hallucinations, and improve response quality**.  

ðŸ“Œ **Example:**  
- Before RLHF: AI gives **unreliable** medical advice.  
- After RLHF: AI **avoids harmful responses** and **follows ethical guidelines**.  

---

### **3. RAG (Retrieval-Augmented Generation) (User Perspective)**  
ðŸ“Œ **Definition:**  
- A technique where an AI model retrieves **external private data** before generating an answer.  
- **Hybrid approach**: Combines **search (retrieval)** and **AI-generated text (generation)**.  

ðŸ“Œ **Why It Matters?**  
- Helps LLMs **answer on private datasets without fine-tuning**.  
- Used in **AI chatbots, business automation, and customer support**.  

ðŸ“Œ **Example:**  
- A company wants an AI chatbot **trained on internal documents** â†’ **RAG enables this without retraining the entire model**.  

---

### **4. Pretraining (Builder Perspective)**  
ðŸ“Œ **Definition:**  
- The process of **training a foundation model from scratch** on massive datasets.  
- Requires **supercomputers, large datasets, and months of computation**.  

ðŸ“Œ **Why It Matters?**  
- Determines the **knowledge** and **capabilities** of an AI model.  
- **Example:** GPT models are pretrained on **internet-scale data** before fine-tuning.  

ðŸ“Œ **Key Steps in Pretraining:**  
1. **Data Collection** â€“ Billions of text, images, code.  
2. **Tokenization** â€“ Converts text into numbers.  
3. **Model Training** â€“ Uses GPUs/TPUs for months.  

---

### **5. Quantization (Builder Perspective)**  
ðŸ“Œ **Definition:**  
- A method to **compress large AI models** so they can run on smaller devices.  
- Reduces **memory usage** and **speeds up inference**.  

ðŸ“Œ **Why It Matters?**  
- Allows **LLMs to run on mobile phones, laptops, and edge devices**.  
- Used in **open-source models like LLaMA, Mistral, and Stable Diffusion**.  

ðŸ“Œ **Example:**  
- GPT-4 **requires cloud GPUs**, but a **quantized LLaMA model** can run on a laptop.  

---

### **6. AI Agents (User Perspective)**  
ðŸ“Œ **Definition:**  
- Autonomous AI systems that **take actions** based on environment and goals.  
- Unlike LLMs (which just generate text), AI Agents can **use tools, browse the web, and automate workflows**.  

ðŸ“Œ **Why It Matters?**  
- Used in **customer service, workflow automation, and personal AI assistants**.  
- **Example:** AI agents **book flights, manage calendars, and analyze stock markets**.  

ðŸ“Œ **Popular AI Agents:**  
- **Auto-GPT, BabyAGI, AgentGPT** â€“ Work **autonomously** with minimal input.  

---

### **7. Vector Databases (User Perspective)**  
ðŸ“Œ **Definition:**  
- Specialized databases that **store and search high-dimensional AI embeddings**.  
- Used for **semantic search, recommendation systems, and RAG-based AI applications**.  

ðŸ“Œ **Why It Matters?**  
- AI doesnâ€™t understand raw text; it converts it into **vectors** (numerical representations).  
- **Vector databases efficiently retrieve similar data points.**  

ðŸ“Œ **Examples of Vector Databases:**  
- **FAISS (by Facebook), Pinecone, Weaviate, ChromaDB.**  

ðŸ“Œ **Use Case:**  
- **Spotify** uses vector databases to **recommend songs based on similarity**.  

---

### **8. Fine-Tuning (User & Builder Perspective)**  
ðŸ“Œ **Definition:**  
- The process of **training a foundation model on specific datasets** to adapt it to a particular use case.  
- Unlike pretraining, fine-tuning **requires much less data and computation**.  

ðŸ“Œ **Why It Matters?**  
- **User Perspective:** Fine-tune models on **custom business data**.  
- **Builder Perspective:** Improve general-purpose models for **specific industries**.  

ðŸ“Œ **Example:**  
- A legal firm fine-tunes **GPT-4 on case laws** to build an **AI legal assistant**.

### **2ï¸âƒ£ Types of Transformers**  
Transformers are classified based on their architecture:  

ðŸ“Œ **Encoder-Only Models (BERT, RoBERTa, ALBERT)**  
   - Good for **understanding text** (e.g., sentiment analysis).  
   - Trained using **Masked Language Modeling (MLM)**.  

ðŸ“Œ **Decoder-Only Models (GPT-3, GPT-4, LLaMA, Mistral)**  
   - Good for **text generation** (e.g., chatbots, writing code).  
   - Uses **Causal Language Modeling (CLM)** (predicts next word).  

ðŸ“Œ **Encoder-Decoder Models (T5, BART, mT5, FLAN-T5)**  
   - Good for **translation, summarization, Q&A** tasks.  
   - Uses **Seq2Seq training** (encoder encodes input, decoder generates output).

### **3ï¸âƒ£ Pretraining a Model from Scratch**  
Pretraining is **expensive** but essential for foundation models.  

ðŸ“Œ **Key Steps:**  
âœ… **Training Objective:**  
   - **Causal LM (CLM)** â†’ For **decoder-only models** like GPT.  
   - **Masked LM (MLM)** â†’ For **encoder-only models** like BERT.  
   - **Seq2Seq Training** â†’ For **encoder-decoder models** like T5.  

âœ… **Tokenization Strategies:**  
   - **Byte-Pair Encoding (BPE)** â†’ Used by GPT models.

âœ… **Training Strategies:**  
   - **Data Parallelism** â†’ Splits data across multiple GPUs.

   - **Mixed Precision Training** â†’ Uses FP16 instead of FP32 for faster training.  

âœ… **Handling Challenges:**

âœ… **Evaluation:**

# **3ï¸âƒ£ Agents â€“ LLMs That Can Take Actions**
- Traditional LLM chatbots **only generate text**.  
- Agents **can take actions** (e.g., book tickets, run scripts).  
- **Example:** Ask a chatbot to book a flight, and it interacts with an API.  

**How do they work?**  
âœ… LLM + **Tool Use** (APIs, browser, Python execution).  
âœ… Example frameworks: **LangChain Agents, AutoGPT, BabyAGI**.  

ðŸ“Œ **Example of an AI agent:**  
```python
from langchain.agents import initialize_agent
from langchain.tools import Tool

def book_flight(destination):
    return f"Flight booked to {destination}!"

agent = initialize_agent(
    tools=[Tool(name="Flight Booker", func=book_flight)],
    llm=OpenAI(api_key="your-api-key")
)

print(agent.run("Book a flight to Paris."))
```
